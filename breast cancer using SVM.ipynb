```python
# importing the dataset
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
cancer = load_breast_cancer()
list(cancer.target_names)
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = cancer.target
df
```

```python
#Describe
df.describe()
```

# Pairplot for visualizing relationships between features

```python
sns.pairplot(df, hue='target', vars=cancer.feature_names[:5])
plt.suptitle("Pairplot of Breast Cancer Dataset Features (First 5)", y=1.02)
plt.show()
```

# Heatmap to visualize correlation between features

```python
plt.figure(figsize=(14, 10))  # Increase the figure size
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Breast Cancer Dataset Features')
plt.show()
```

# Boxplot for feature comparison

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='target', y='mean radius', data=df)
plt.title('Comparison of Mean Radius by Target Class')
plt.xlabel('Target Class')
plt.ylabel('Mean Radius')
plt.show()
```

# Violin plot for feature comparison

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='target', y='mean texture', data=df)
plt.title('Comparison of Mean Texture by Target Class')
plt.xlabel('Target Class')
plt.ylabel('Mean Texture')
plt.show()
```

# Histograms for feature distribution

```python
plt.figure(figsize=(12, 8))
for i, feature in enumerate(cancer.feature_names[:5]):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[feature], kde=True, color='skyblue', bins=20)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# Swarm plot for feature comparison

# Violin plot with split for feature comparison

# Countplot for target class distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=df)
plt.title('Distribution of Target Classes')
plt.xlabel('Target Class')
plt.ylabel('Count')
plt.show()
```

```python
#Splice Dataset into X and Y
X=cancer
y=cancer.target_names
y
```

```python
# splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target, stratify=cancer.target,test_size=0.25 ,random_state=0)
print(X_train)

```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

```python
#Training the SVM on the Training set
from sklearn.svm import SVC
svm= SVC (random_state=0)
svm.fit(X_train, y_train)
print(y_train)
```

```python
#Predicting the test set result
y_pred = svm.predict(X_test)
print("Predicting the test set result for SVM: ",y_pred)
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix for SVM:")
print(cm)
accuracy=accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM for Breast Cancer : {accuracy}")
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm_df = pd.DataFrame(cm ,index = ['0','1'], columns = ['0','1'])
plt.figure(figsize=(7,5))
sns.heatmap(cm_df, annot=True)
plt.title('SVM Confusion Matrix for Breast Cancer')
plt.ylabel('Actual Class ')
plt.xlabel('Predicted Class ')
plt.show()
```

```python
#Evaluating Precision, Recall and f-measure
from sklearn.metrics import precision_score,recall_score
precision=precision_score(y_test,y_pred)
print("Precision: ",precision*100)
Recall=recall_score(y_test,y_pred)
print("Recall:    ",Recall*100)
f1score=2*(precision*Recall)/(precision+Recall)
print("F1-score:",f1score*100)
```

```python
#Measure The Error Rate of Model
from sklearn.metrics import mean_squared_error,mean_absolute_error
MSE=mean_squared_error(y_test, y_pred)
MAR=mean_absolute_error(y_test, y_pred)
print(f'Mean Sqaure Error: {MSE}')
print(f'Mean Absolute Error: {MAR}')
```
